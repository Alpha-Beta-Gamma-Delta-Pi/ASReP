```python
def proper_form(data):
    s,e = 0,0
    all_data = {}
    i,j = '',0

    while (e<len(data)):
        e = data.index(':',s+1) if ':' in data[s+1:] else len(data)
        s += 2
        dt = (data[s:e])
        # print(dt,s,e)
        if (dt[:3] == "200"):
            # print("-------")
            i = dt[4:]
            all_data[i] = {}
            # print(dt[4:])
        # elif (e==-1):
        #     print("-------")
        elif (dt[0] == '{'):
            # print(",,,,,")
            # print(dt[1:])
            j = dt[2:-1]
            # print(j)
            all_data[i][j] = ""
        elif (dt[:5] == "array"):
            # print(".....")
            e = dt.index(']')
            dt = dt[7:e] #.strip()
            # print([float(i) for i in dt.split(",")])
            all_data[i][j] = [float(i) for i in dt.split(",")]
            e += 2
            # s=e
            e += s
            # print(data[s:e],s,e)
            # s += e
        elif "mrr" in dt:
            # print("*****")
            # print(dt[:dt.index(',')])
            all_data[i][j] = float(dt[:dt.index(',')])
            e += 2
            j = "mrr"
            all_data[i][j] = ""
            # print("mrr")
        elif dt == "epoch":
            print("!!!!!!!!!!!!")
            # continue
        elif e == len(data):   #"\n\nepoch" in dt:
            val = "0." + dt[:-1]
            # print(float(val))
            all_data[i][j] = float(val)
        elif "\nepoch" in dt:
            val = "0." + dt[:dt.index('}')]
            # print(float(val))
            all_data[i][j] = float(val)
        else:
            # print("#########")
            j = dt[1:-1]
            # print(j)
            all_data[i][j] = 0

        s = e
        # print("@@@@@@@@@@@@@@@@@@@@@")
        # print(all_data)
    return all_data
```

```python
original_results = proper_form(data)
```
